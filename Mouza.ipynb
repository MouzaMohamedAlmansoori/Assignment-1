```python
class Flight:
    def __init__(self, flight_number, date, departure_time, arrival_time, terminal, gate, start_city, destination_city):
         """
         initilize the flight object wuth different Parameters
         
         Parametere:
         
         flight_number(str):passing the flight number 
         date(date):passing the date of the flight
         departure_time:passing the departure time
         arrival_time:passing the arrival time
         terminal:passsig the terminal of the departure
         gate:the gate assign for the departure
         start_city:send the name of the departure city
         destination_city:send the name of the arrival city
         
         Raises Error:
         If the user enter the incomplete flight details
          """
        
        if not all([flight_number, date, departure_time, arrival_time, terminal, gate, start_city, destination_city]):
            raise ValueError("Incomplete Flight Details")
        self.flight_number = flight_number
        self.date = date
        self.departure_time = departure_time
        self.arrival_time = arrival_time
        self.terminal = terminal
        self.gate = gate
        self.start_city = start_city
        self.destination_city = destination_city




class BoardingPass:               # Create the class of the Boarding Pass
    
     """
         initilize the boarding pass object with different Parameters
         
         Parametere:
         
         passenger:passing the passenger information which is associated with the boarding pass 
         flight:passing the detail of the flight which is associated with the boarding pass
         seat_number:passing the seat number which is  associated to the pessenger
         gate:the gate assign for the departure
         boarding_time:Passing the time at which the boarding starts
         
        
          """
    def __init__(self, passenger, flight, seat_number, gate, boarding_time):
        self.passenger = passenger
        self.flight = flight
        self.seat_number = seat_number
        self.gate = gate
        self.boarding_time = boarding_time





class Passenger:                # Create the class of the Passenger
    
      """
         initilize the passenger object with different Parameters
         
         Parametere:
         
        name:passing the passenger name 
         ticket_number:passing the ticket number
         seat_number:passing the seat number which is  associated to the pessenger
       
          """
        
    def __init__(self, name, ticket_number, seat_number):
        self.name = name
        self.ticket_number = ticket_number
        self.seat_number = seat_number
        self.boarding_pass = None

    def set_boarding_pass(self, boarding_pass):
      
    """
         set the boarding pass to the passenger
         
         Parametere:
         
         boarding_pass = (boarding_pass): The boarding pass objects to be assign
       
          """
        self.boarding_pass = boarding_pass

    def get_boarding_pass(self):
         """
        Retrieve the data  of the boarding pass to the passenger
         
         Return:
         
         boarding_pass The boarding_pass object is assign to the pessenger
       
          """
        return self.boarding_pass



class Airline:           # Create the class of the Airline
     """
         initilize the Airline object with different Parameters
         
         Parametere:
         
        name:passing the  name of the airline 
        
       
          """
        
    def __init__(self, name):
        
        self.name = name
        self.flights = []

    def add_flight(self, flight):
        
        """
         set the add teh flight in the system
         
    
       
          """
        self.flights.append(flight)


```

```python
# Create Passenger
passenger = Passenger("Mouza", "5A6BCD78", "09A")

# Create Flight
flight = Flight("NA4321", "06 DEC 20", "11:40", "13:30", "ORD", "03", "CHICAGO", "NEW York")

# Create Boarding Pass
boarding_pass = BoardingPass(passenger, flight, "09A", "03", "11:20")

passenger.set_boarding_pass(boarding_pass)     # Passenger associates with Boarding Pass
```
